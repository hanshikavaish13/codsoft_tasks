```python
import tkinter as tk
from tkinter import messagebox
import json
import os

FILE_NAME = "contacts.json"
```

```python
#Load Contacts 
def load_contacts():
    if os.path.exists(FILE_NAME):
        with open(FILE_NAME, "r") as file:
            return json.load(file)
    return []
```

```python
#Save Contacts
def save_contacts():
    with open(FILE_NAME, "w") as file:
        json.dump(contacts, file, indent=4)
```

```python
#Refresh List
def refresh_list():
    contact_list.delete(0, tk.END)
    for contact in contacts:
        contact_list.insert(tk.END,f"{contact['name']} - {contact['phone']}")
```

```python
#Add Contact
def add_contact():
    name = name_entry.get()
    phone = phone_entry.get()
    email = email_entry.get()
    address = address_entry.get()
    if name == "" or phone == "":
        messagebox.showwarning("Warning","Name and Phone Number are required!")
        return
    contact = {"name": name,"phone": phone,"email": email,"address": address}
    contacts.append(contact)
    save_contacts()
    refresh_list()
    clear_fields()
    messagebox.showinfo("Success", "Contact Added Successfully!")
```

```python
#Search Contact
def search_contact():
    query = search_entry.get().lower()
    contact_list.delete(0, tk.END)
    for contact in contacts:
        if query in contact["name"].lower() or query in contact["phone"]:
            contact_list.insert(tk.END,f"{contact['name']} - {contact['phone']}")
```

```python
#Select Contact
def select_contact(event):
    try:
        index = contact_list.curselection()[0]
        selected = contact_list.get(index)
        for contact in contacts:
            if f"{contact['name']} - {contact['phone']}" == selected:
                name_entry.delete(0, tk.END)
                phone_entry.delete(0, tk.END)
                email_entry.delete(0, tk.END)
                address_entry.delete(0, tk.END)
                name_entry.insert(0, contact["name"])
                phone_entry.insert(0, contact["phone"])
                email_entry.insert(0, contact["email"])
                address_entry.insert(0, contact["address"])
    except:
        pass
```

```python
#Update Contact
def update_contact():
    selected = contact_list.curselection()
    if not selected:
        messagebox.showwarning("Warning","Select a contact to update!")
        return
    old_text = contact_list.get(selected[0])
    for contact in contacts:
        if f"{contact['name']} - {contact['phone']}" == old_text:
            contact["name"] = name_entry.get()
            contact["phone"] = phone_entry.get()
            contact["email"] = email_entry.get()
            contact["address"] = address_entry.get()
            break
    save_contacts()
    refresh_list()
    clear_fields()
    messagebox.showinfo("Success", "Contact Updated Successfully!")
```

```python
#Delete Contact
def delete_contact():
    selected = contact_list.curselection()

    if not selected:
        messagebox.showwarning("Warning","Select a contact to delete!")
        return
    selected_text = contact_list.get(selected[0])
    for contact in contacts:
        if f"{contact['name']} - {contact['phone']}" == selected_text:
            contacts.remove(contact)
            break
    save_contacts()
    refresh_list()
    clear_fields()
    messagebox.showinfo("Success", "Contact Deleted Successfully!")
```

```python
#Clear Fields
def clear_fields():
    name_entry.delete(0, tk.END)
    phone_entry.delete(0, tk.END)
    email_entry.delete(0, tk.END)
    address_entry.delete(0, tk.END)
```

```python
# GUI
import tkinter as tk

root = tk.Tk()
root.title("📒 Contact Book")
root.geometry("700x650")
root.configure(bg="#0F172A")
root.resizable(False, False)

contacts = load_contacts()

#Colors
BG = "#0F172A"
CARD = "#1E293B"
TEXT = "#F8FAFC"
ENTRY_BG = "#E2E8F0"

BLUE = "#3B82F6"
GREEN = "#22C55E"
ORANGE = "#F59E0B"
RED = "#EF4444"
PURPLE = "#8B5CF6"

#Heading
title = tk.Label(
    root,
    text="📒 CONTACT BOOK",
    bg=BG,
    fg="white",
    font=("Segoe UI", 24, "bold")
)
title.pack(pady=15)

#Main Card
main_frame = tk.Frame(root, bg=CARD, bd=0)
main_frame.pack(padx=20, pady=10, fill="both", expand=True)

#Labels & Entries
label_font = ("Segoe UI", 10, "bold")
entry_font = ("Segoe UI", 11)

tk.Label(main_frame, text="Name", bg=CARD, fg=TEXT,
         font=label_font).grid(row=0, column=0, padx=15, pady=10, sticky="w")

name_entry = tk.Entry(
    main_frame,
    font=entry_font,
    width=35,
    bg=ENTRY_BG,
    relief="flat"
)
name_entry.grid(row=0, column=1, pady=10)

tk.Label(main_frame, text="Phone Number", bg=CARD, fg=TEXT,
         font=label_font).grid(row=1, column=0, padx=15, pady=10, sticky="w")

phone_entry = tk.Entry(
    main_frame,
    font=entry_font,
    width=35,
    bg=ENTRY_BG,
    relief="flat"
)
phone_entry.grid(row=1, column=1, pady=10)

tk.Label(main_frame, text="Email", bg=CARD, fg=TEXT,
         font=label_font).grid(row=2, column=0, padx=15, pady=10, sticky="w")

email_entry = tk.Entry(
    main_frame,
    font=entry_font,
    width=35,
    bg=ENTRY_BG,
    relief="flat"
)
email_entry.grid(row=2, column=1, pady=10)

tk.Label(main_frame, text="Address", bg=CARD, fg=TEXT,
         font=label_font).grid(row=3, column=0, padx=15, pady=10, sticky="w")

address_entry = tk.Entry(
    main_frame,
    font=entry_font,
    width=35,
    bg=ENTRY_BG,
    relief="flat"
)
address_entry.grid(row=3, column=1, pady=10)

#Search
tk.Label(main_frame, text="Search", bg=CARD, fg=TEXT,
         font=label_font).grid(row=4, column=0, padx=15, pady=15, sticky="w")

search_entry = tk.Entry(
    main_frame,
    font=entry_font,
    width=25,
    bg=ENTRY_BG,
    relief="flat"
)
search_entry.grid(row=4, column=1, sticky="w")

search_btn = tk.Button(
    main_frame,
    text="🔍 Search",
    command=search_contact,
    bg=PURPLE,
    fg="white",
    activebackground=PURPLE,
    relief="flat",
    cursor="hand2",
    font=("Segoe UI", 10, "bold"),
    padx=10
)
search_btn.grid(row=4, column=1, sticky="e")

#Buttons
button_frame = tk.Frame(main_frame, bg=CARD)
button_frame.grid(row=5, column=0, columnspan=2, pady=20)

btn_style = {
    "font": ("Segoe UI", 10, "bold"),
    "fg": "white",
    "relief": "flat",
    "cursor": "hand2",
    "width": 12,
    "height": 1
}

tk.Button(
    button_frame,
    text="➕ Add",
    command=add_contact,
    bg=GREEN,
    **btn_style
).grid(row=0, column=0, padx=6)

tk.Button(
    button_frame,
    text="✏ Update",
    command=update_contact,
    bg=ORANGE,
    **btn_style
).grid(row=0, column=1, padx=6)

tk.Button(
    button_frame,
    text="🗑 Delete",
    command=delete_contact,
    bg=RED,
    **btn_style
).grid(row=0, column=2, padx=6)

tk.Button(
    button_frame,
    text="🧹 Clear",
    command=clear_fields,
    bg=BLUE,
    **btn_style
).grid(row=0, column=3, padx=6)

#Contact List
list_frame = tk.Frame(main_frame, bg=CARD)
list_frame.grid(row=6, column=0, columnspan=2, padx=15, pady=15)

scrollbar = tk.Scrollbar(list_frame)

contact_list = tk.Listbox(
    list_frame,
    width=60,
    height=12,
    font=("Segoe UI", 11),
    bg="white",
    fg="#1F2937",
    selectbackground=BLUE,
    selectforeground="white",
    relief="flat",
    yscrollcommand=scrollbar.set
)

scrollbar.config(command=contact_list.yview)

contact_list.pack(side="left")
scrollbar.pack(side="right", fill="y")

contact_list.bind("<<ListboxSelect>>", select_contact)

#Load Contacts
refresh_list()

root.mainloop()
```
