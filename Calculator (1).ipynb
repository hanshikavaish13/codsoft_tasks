```python
import tkinter as tk
from tkinter import ttk
import math

#Calculator Function
def calculate():
    try:
        num1 = float(entry1.get())
    except ValueError:
        result.config(text="Enter a valid First Number!", fg="#EF4444")
        return

    operation = operation_box.get()

    if operation != "√ Square Root":
        try:
            num2 = float(entry2.get())
        except ValueError:
            result.config(text="Enter a valid Second Number!", fg="#EF4444")
            return

    try:
        if operation == "➕ Addition":
            ans = num1 + num2

        elif operation == "➖ Subtraction":
            ans = num1 - num2

        elif operation == "✖ Multiplication":
            ans = num1 * num2

        elif operation == "➗ Division":
            if num2 == 0:
                result.config(text="Cannot divide by zero!", fg="#EF4444")
                return
            ans = num1 / num2

        elif operation == "% Modulus":
            ans = num1 % num2

        elif operation == "Power":
            ans = num1 ** num2

        elif operation == "Floor Division":
            ans = num1 // num2

        elif operation == "√ Square Root":
            if num1 < 0:
                result.config(text="Negative number!", fg="#EF4444")
                return
            ans = math.sqrt(num1)

        else:
            result.config(text="Select an operation!", fg="#EF4444")
            return

        result.config(text=f"Result = {ans}", fg="#22C55E")

    except Exception:
        result.config(text="Calculation Error!", fg="#EF4444")


#Window
root = tk.Tk()
root.title("🧮 Smart Calculator")
root.geometry("600x600")
root.configure(bg="#0F172A")
root.resizable(False, False)

# Colors
BG = "#0F172A"
CARD = "#1E293B"
TEXT = "#F8FAFC"
ENTRY = "#E2E8F0"
BLUE = "#3B82F6"

# Heading
title = tk.Label(
    root,
    text="🧮 SMART CALCULATOR",
    bg=BG,
    fg="white",
    font=("Segoe UI", 22, "bold")
)
title.pack(pady=20)

# Card
frame = tk.Frame(root, bg=CARD)
frame.pack(padx=20, pady=10, fill="both", expand=True)

# First Number
tk.Label(
    frame,
    text="First Number",
    bg=CARD,
    fg=TEXT,
    font=("Segoe UI", 11, "bold")
).pack(anchor="w", padx=20, pady=(20,5))

entry1 = tk.Entry(
    frame,
    font=("Segoe UI",11),
    bg=ENTRY,
    relief="flat"
)
entry1.pack(fill="x", padx=20)

# Second Number
tk.Label(
    frame,
    text="Second Number",
    bg=CARD,
    fg=TEXT,
    font=("Segoe UI",11,"bold")
).pack(anchor="w", padx=20, pady=(20,5))

entry2 = tk.Entry(
    frame,
    font=("Segoe UI",11),
    bg=ENTRY,
    relief="flat"
)
entry2.pack(fill="x", padx=20)

# Operation
tk.Label(
    frame,
    text="Select Operation",
    bg=CARD,
    fg=TEXT,
    font=("Segoe UI",11,"bold")
).pack(anchor="w", padx=20, pady=(20,5))

operation_box = ttk.Combobox(
    frame,
    state="readonly",
    font=("Segoe UI",11),
    values=[
        "➕ Addition",
        "➖ Subtraction",
        "✖ Multiplication",
        "➗ Division",
        "% Modulus",
        "Power",
        "Floor Division",
        "√ Square Root"
    ]
)

operation_box.current(0)
operation_box.pack(fill="x", padx=20)

# Calculate Button
calculate_btn = tk.Button(
    frame,
    text="Calculate",
    command=calculate,
    bg=BLUE,
    fg="white",
    relief="flat",
    font=("Segoe UI",12,"bold"),
    cursor="hand2",
    width=20,
    pady=8
)
calculate_btn.pack(pady=30)

# Result
result = tk.Label(
    frame,
    text="Result =",
    bg=CARD,
    fg="white",
    font=("Segoe UI",18,"bold")
)
result.pack(pady=20)

# Footer
footer = tk.Label(
    root,
    text="Thank you for using Smart Calculator!",
    bg=BG,
    fg="#94A3B8",
    font=("Segoe UI",10)
)
footer.pack(pady=10)

root.mainloop()
```
