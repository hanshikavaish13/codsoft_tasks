```python
import tkinter as tk
import random
choices = ["Rock", "Paper", "Scissors"]

user_score = 0
computer_score = 0
```

```python
#Game Logic Function
def play(user_choice):
    global user_score, computer_score

    computer_choice = random.choice(choices)

    if user_choice == computer_choice:
        result = "It's a Tie!"
    elif (
        (user_choice == "Rock" and computer_choice == "Scissors") or
        (user_choice == "Paper" and computer_choice == "Rock") or
        (user_choice == "Scissors" and computer_choice == "Paper")
    ):
        result = "You Win!"
        user_score += 1
    else:
        result = "Computer Wins!"
        computer_score += 1

    user_choice_label.config(text=f"Your Choice: {user_choice}")
    computer_choice_label.config(text=f"Computer Choice: {computer_choice}")
    result_label.config(text=result)

    score_label.config(
        text=f"Your Score: {user_score}    Computer Score: {computer_score}"
    )
```

```python
#Reset Game Function
def reset_game():
    global user_score, computer_score

    user_score = 0
    computer_score = 0

    user_choice_label.config(text="Your Choice:")
    computer_choice_label.config(text="Computer Choice:")
    result_label.config(text="Choose Rock, Paper, or Scissors")
    score_label.config(text="Your Score: 0    Computer Score: 0")
```

```python
#GUI 

root = tk.Tk()
root.title("🎮 Rock Paper Scissors")
root.geometry("650x550")
root.configure(bg="#0F172A")
root.resizable(False, False)

#Colors
BG = "#0F172A"
CARD = "#1E293B"
TEXT = "#F8FAFC"

BLUE = "#3B82F6"
GREEN = "#22C55E"
RED = "#EF4444"
ORANGE = "#F59E0B"
PURPLE = "#8B5CF6"

#Title
title_label = tk.Label(
    root,
    text="🎮 ROCK • PAPER • SCISSORS",
    bg=BG,
    fg="white",
    font=("Segoe UI", 24, "bold")
)
title_label.pack(pady=20)

#Main Card
main_frame = tk.Frame(root, bg=CARD)
main_frame.pack(padx=20, pady=10, fill="both", expand=True)

instruction_label = tk.Label(
    main_frame,
    text="Choose your move",
    bg=CARD,
    fg=TEXT,
    font=("Segoe UI", 14)
)
instruction_label.pack(pady=15)

#Buttons
button_frame = tk.Frame(main_frame, bg=CARD)
button_frame.pack(pady=10)

button_style = {
    "font": ("Segoe UI", 12, "bold"),
    "fg": "white",
    "relief": "flat",
    "cursor": "hand2",
    "width": 12,
    "height": 2
}

rock_btn = tk.Button(
    button_frame,
    text="🪨 Rock",
    bg=BLUE,
    command=lambda: play("Rock"),
    **button_style
)
rock_btn.grid(row=0, column=0, padx=10)

paper_btn = tk.Button(
    button_frame,
    text="📄 Paper",
    bg=GREEN,
    command=lambda: play("Paper"),
    **button_style
)
paper_btn.grid(row=0, column=1, padx=10)

scissors_btn = tk.Button(
    button_frame,
    text="✂️ Scissors",
    bg=ORANGE,
    command=lambda: play("Scissors"),
    **button_style
)
scissors_btn.grid(row=0, column=2, padx=10)

# -------- Result Card --------
result_frame = tk.Frame(
    main_frame,
    bg="#334155",
    padx=20,
    pady=20
)
result_frame.pack(pady=30, fill="x", padx=20)

user_choice_label = tk.Label(
    result_frame,
    text="Your Choice:",
    bg="#334155",
    fg="white",
    font=("Segoe UI", 12)
)
user_choice_label.pack(pady=5)

computer_choice_label = tk.Label(
    result_frame,
    text="Computer Choice:",
    bg="#334155",
    fg="white",
    font=("Segoe UI", 12)
)
computer_choice_label.pack(pady=5)

result_label = tk.Label(
    result_frame,
    text="Choose Rock, Paper or Scissors",
    bg="#334155",
    fg="#FBBF24",
    font=("Segoe UI", 18, "bold")
)
result_label.pack(pady=15)

score_label = tk.Label(
    result_frame,
    text="Your Score: 0    Computer Score: 0",
    bg="#334155",
    fg="white",
    font=("Segoe UI", 13, "bold")
)
score_label.pack(pady=10)

#Reset Button
reset_btn = tk.Button(
    main_frame,
    text="🔄 Reset Game",
    command=reset_game,
    bg=RED,
    fg="white",
    relief="flat",
    cursor="hand2",
    font=("Segoe UI", 12, "bold"),
    width=18,
    height=2
)
reset_btn.pack(pady=20)

root.mainloop()
```
