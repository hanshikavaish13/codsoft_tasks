```python
import tkinter as tk
from tkinter import ttk, messagebox
import pandas as pd
import os
```

```python
filename = "tasks.csv"

if not os.path.exists(filename):
    df = pd.DataFrame(columns=["Task"])
    df.to_csv(filename, index=False)
```

```python
root = tk.Tk()
root.title("📝 To-Do List")
root.geometry("700x550")
root.configure(bg="#0F172A")
root.resizable(False, False)
```

```python
title = tk.Label(
    root,
    text="📝 TO-DO LIST",
    bg="#0F172A",
    fg="white",
    font=("Segoe UI",22,"bold")
)

title.pack(pady=15)
```

```python
task_entry = tk.Entry(
    root,
    font=("Segoe UI",12),
    width=35
)

task_entry.pack(pady=10)
```

```python
tree = ttk.Treeview(root, columns=("Task"), show="headings")

tree.heading("Task", text="Task")

tree.pack(fill="both", expand=True, padx=20, pady=20)
```

```python
def refresh():

    tree.delete(*tree.get_children())

    df = pd.read_csv(filename)

    for i,row in df.iterrows():
        tree.insert("", tk.END, values=(row["Task"],))
```

```python
def add_task():

    task = task_entry.get()

    if task == "":
        messagebox.showwarning("Warning","Enter Task")
        return

    df = pd.read_csv(filename)

    df.loc[len(df)] = [task]

    df.to_csv(filename,index=False)

    task_entry.delete(0,tk.END)

    refresh()
```

```python
def delete_task():

    selected = tree.selection()

    if not selected:
        messagebox.showwarning("Warning","Select a Task")
        return

    item = tree.item(selected[0])

    task = item["values"][0]

    df = pd.read_csv(filename)

    df = df[df["Task"] != task]

    df.to_csv(filename,index=False)

    refresh()
```

```python
def update_task():

    selected = tree.selection()

    if not selected:
        messagebox.showwarning("Warning","Select a Task")
        return

    new_task = task_entry.get()

    if new_task == "":
        messagebox.showwarning("Warning","Enter New Task")
        return

    item = tree.item(selected[0])

    old_task = item["values"][0]

    df = pd.read_csv(filename)

    df.loc[df["Task"] == old_task,"Task"] = new_task

    df.to_csv(filename,index=False)

    task_entry.delete(0,tk.END)

    refresh()
```

```python
def load_task(event):

    selected = tree.selection()

    if selected:

        item = tree.item(selected[0])

        task_entry.delete(0,tk.END)

        task_entry.insert(0,item["values"][0])

tree.bind("<<TreeviewSelect>>", load_task)
```

```python
def clear_task():

    task_entry.delete(0,tk.END)
```

```python
button_frame = tk.Frame(root,bg="#0F172A")

button_frame.pack(pady=10)

tk.Button(button_frame,text="➕ Add",bg="#22C55E",fg="white",
command=add_task,width=10).grid(row=0,column=0,padx=5)

tk.Button(button_frame,text="✏ Update",bg="#F59E0B",fg="white",
command=update_task,width=10).grid(row=0,column=1,padx=5)

tk.Button(button_frame,text="🗑 Delete",bg="#EF4444",fg="white",
command=delete_task,width=10).grid(row=0,column=2,padx=5)

tk.Button(button_frame,text="🧹 Clear",bg="#3B82F6",fg="white",
command=clear_task,width=10).grid(row=0,column=3,padx=5)

tk.Button(button_frame,text="🔄 Refresh",bg="#8B5CF6",fg="white",
command=refresh,width=10).grid(row=0,column=4,padx=5)
```

```python
refresh()

root.mainloop()
```
